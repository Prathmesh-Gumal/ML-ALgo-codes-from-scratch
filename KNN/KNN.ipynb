```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df =pd.read_csv("/content/data.csv")
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 32'],inplace=True)
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
#The iloc[:,1:] was changed to  df.drop(columns=['diagnosis']). This is done to explicitly remove the diagnosis column which had the string values.
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df.iloc[:,0],test_size=0.2, random_state = 2)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
!pip install scikit-learn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
